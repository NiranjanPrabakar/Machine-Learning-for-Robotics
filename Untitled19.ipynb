```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/rainfall in india 1901-2015 (1).csv")

# Preprocessing: Select Tamil Nadu data
df_tn = df[df['SUBDIVISION'] == 'TAMIL NADU']
df_tn = df_tn[['YEAR', 'ANNUAL']].reset_index(drop=True)

# Handle missing values (if any)
df_tn.fillna(df_tn.mean(), inplace=True)

# Normalize data
scaler = MinMaxScaler()
df_tn['ANNUAL'] = scaler.fit_transform(df_tn[['ANNUAL']])

# Prepare dataset for LSTM
X, y = [], []
seq_length = 10  # Using last 10 years to predict the next year
for i in range(len(df_tn) - seq_length):
    X.append(df_tn['ANNUAL'].iloc[i:i+seq_length].values)
    y.append(df_tn['ANNUAL'].iloc[i+seq_length])
X, y = np.array(X), np.array(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Build LSTM Model
model = Sequential([
    LSTM(50, activation='relu', return_sequences=True, input_shape=(seq_length, 1)),
    Dropout(0.2),
    LSTM(50, activation='relu'),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=8, validation_data=(X_test, y_test))

# Predict for 2016-2025
future_years = np.arange(2016, 2026)
predictions = []

last_sequence = X[-1]  # Start with last known sequence
for year in future_years:
    pred = model.predict(last_sequence.reshape(1, seq_length, 1))[0, 0]
    predictions.append(pred)
    last_sequence = np.roll(last_sequence, -1)
    last_sequence[-1] = pred

# Convert predictions back to original scale
predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, 1)).flatten()

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(df_tn['YEAR'], scaler.inverse_transform(df_tn[['ANNUAL']]), label='Actual', marker='o')
plt.plot(future_years, predictions, label='Predicted', marker='s', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Annual Rainfall (mm)")
plt.title("Tamil Nadu Rainfall Prediction (2016-2025)")
plt.legend()
plt.grid()
plt.show()

# Display predictions
tn_predictions_df = pd.DataFrame({"Year": future_years, "Predicted Rainfall": predictions})
print(tn_predictions_df)
```
